#2. Apply appropriate mask on an image to perform Box blurring..

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
import cv2 as cv
def main():   #Function defination
    
    img = cv2.imread("/content/imgc.JPG")
    #img = cv2.imread("/content/imgc_color.tif")
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    rows, cols, chan = img.shape
    #rows, cols= img.shape
    p = 0.25

    output = np.zeros(img.shape, np.uint8)

    for i in range(rows):
      for j in range(cols):
        r = random.random()
        if r < p/2:
          # pepper sprinkled 
          output[i][j] = [0,0,0]
        elif r < p:                 #P/2 <r < P
          # salt sprinkled 
          output[i][j] = [255,255,255]
        else:
          output[i][j] = img[i][j]    # r>=p
 
    plt.subplot(121)
    plt.imshow(img)
    plt.title('Original Image')
    plt.subplot(122)
    plt.imshow(output)
    plt.title('Image with Salt and Pepper Noise')
    plt.show()

if __name__ == "__main__":
    main()   # calling
```
